# There aren't a lot of missing values. Since all the data is normally distributed, we will impute using median for numeric and mode for categorical values.

```python
# Imputation using median and mode
for x in df.columns:
    if df[x].dtype in ['float64','int64']:
        df[x] = df[x].fillna(df[x].median())
    else:
        df[x] = df[x].fillna(df[x].mode()[0])

df.isnull().sum()
```

# XGBoost

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.coded.drop('y', axis = 1)
y = df.coded['y']

# Test train split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=100)

# Data scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_testaled = scaler.fit_transform(X_test)
```

```python
# Hyperparameter tuning

from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=100)

#  parameter grid for random search
param_distributions = {
    'eta': [0.01, 0.05, 0.1, 0.3],  # learning rate
    'max_depth': [3, 5, 7, 10],      # maximum tree depth
    'subsample': [0.6, 0.8, 1.0],    # row subsampling
    'colsample_bytree': [0.6, 0.8, 1.0],  # column subsampling
    'min_child_weight': [1, 5, 10],  # minimum child weight
    'lambda': [0.01, 0.1, 1],        # L2 regularization
    'alpha': [0.01, 0.1, 1],         # L1 regularization
    'n_estimators': [1000], # Number of trees constant, we are not tuning it
    
}

# setting up RandomSearch

random_search = RandomizedSearchCV(estimator = xgb, param_distributions = param_distributions, n_iter = 50,cv=5, scoring = 'roc_auc', verbose= 1)

random_search.fit(X_train_scaled, y_train)

# print the best hyperparameters
random_search.best_params_
```

```python
random_search.best_estimator_
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import numpy as np

# best model from RandomizedSearchCV
best_xgb = random_search.best_estimator_

# Initialize StratifiedKFold for cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)

# To store cross-validation metrics
cv_accuracies = []
cv_auc_scores = []

# Perform cross-validation
for train_idx, val_idx in kf.split(X_train_scaled, y_train):
    # Split the data into training and validation sets
    X_train_cv, X_val_cv = X_train_scaled[train_idx], X_train_scaled[val_idx]
    y_train_cv, y_val_cv = y_train.iloc[train_idx], y_train.iloc[val_idx]

    # Train the model on the training fold
    best_xgb.fit(X_train_cv, y_train_cv)

    # Evaluate accuracy and AUC on the validation fold
    val_accuracy = best_xgb.score(X_val_cv, y_val_cv)
    y_val_probs = best_xgb.predict_proba(X_val_cv)[:, 1]  # Get probabilities for ROC/AUC
    val_auc = roc_auc_score(y_val_cv, y_val_probs)

    # Append metrics
    cv_accuracies.append(val_accuracy)
    cv_auc_scores.append(val_auc)

# Print Cross-Validation Results
print(f"Cross-Validation Accuracy: {np.mean(cv_accuracies):.4f} ± {np.std(cv_accuracies):.4f}")
print(f"Cross-Validation AUC: {np.mean(cv_auc_scores):.4f} ± {np.std(cv_auc_scores):.4f}")
```

```python
# Print all fold results
print("Accuracy Scores for each fold:", cv_accuracies)
print("AUC Scores for each fold:", cv_auc_scores)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Predict probabilities on the test set
y_test_probs = best_xgb.predict_proba(X_testaled)[:, 1] # typo for test_scaled on the top

# ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_test, y_test_probs)
roc_auc = roc_auc_score(y_test, y_test_probs)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})', linewidth=2)
plt.plot([0, 1], [0, 1], 'k--', linewidth=2)
plt.title('ROC Curve(cross-validated)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()

print(f"Test AUC: {roc_auc:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Prediction probabilities
y_test_probs = best_xgb.predict_proba(X_testaled)[:, 1]

# custom threshold
threshold = 0.75  
y_test_pred = (y_test_probs >= threshold).astype(int)

# Confusion matrix
cm = confusion_matrix(y_test, y_test_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_xgb.classes_)
disp.plot(cmap='viridis', values_format='d')
plt.title(f'Confusion Matrix (Threshold = {threshold})')
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, confusion_matrix
import numpy as np

# Calculate scale_pos_weight
negative_samples = (y_train == 0).sum()
positive_samples = (y_train == 1).sum()
scale_pos_weight = (12.50 / 3.50) * (negative_samples / positive_samples)

# Initialize StratifiedKFold for cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)

# To store cross-validation metrics
cv_accuracies = []
cv_auc_scores = []
monetary_losses = []

# Perform cross-validation
for train_idx, val_idx in kf.split(X_train_scaled, y_train):
    # Split the data into training and validation sets
    X_train_cv, X_val_cv = X_train_scaled[train_idx], X_train_scaled[val_idx]
    y_train_cv, y_val_cv = y_train.iloc[train_idx], y_train.iloc[val_idx]

    # Initialize the XGBoost model with class weight
    weighted_xgb = XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        scale_pos_weight=scale_pos_weight,
        random_state=42
    )

    # Train the model on the training fold
    weighted_xgb.fit(X_train_cv, y_train_cv)

    # Predict probabilities on the validation fold
    y_val_probs = weighted_xgb.predict_proba(X_val_cv)[:, 1]
    y_val_pred = (y_val_probs >= 0.75).astype(int)  # Adjust threshold here

    # Calculate accuracy and AUC
    val_accuracy = (y_val_pred == y_val_cv).mean()
    val_auc = roc_auc_score(y_val_cv, y_val_probs)

    # Append metrics
    cv_accuracies.append(val_accuracy)
    cv_auc_scores.append(val_auc)

    # Calculate confusion matrix for monetary loss
    cm = confusion_matrix(y_val_cv, y_val_pred)
    FP = cm[0, 1]  # False Positives
    FN = cm[1, 0]  # False Negatives

    # Calculate monetary loss for this fold
    fold_loss = (12.50 * FP) + (3.50 * FN)
    monetary_losses.append(fold_loss)

# Print Cross-Validation Results
print(f"Cross-Validation Accuracy: {np.mean(cv_accuracies):.4f} ± {np.std(cv_accuracies):.4f}")
print(f"Cross-Validation AUC: {np.mean(cv_auc_scores):.4f} ± {np.std(cv_auc_scores):.4f}")
print(f"Cross-Validation Monetary Loss: {np.mean(monetary_losses):.2f} ± {np.std(monetary_losses):.2f}")
```
